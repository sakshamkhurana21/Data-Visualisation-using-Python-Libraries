```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
u=0
sigma=1
vals=u+sigma*np.random.randn(100)
plt.hist(vals,10)
plt.show()
```

```python
x=vals
y=np.zeros(x.shape)
plt.scatter(x,y)
plt.show()
```

# # Multivariate Normal Distribuion

```python
mean=np.array([0.0,0.0])
cov=np.array([[1,0],[0,1]])
dist=np.random.multivariate_normal(mean,cov,500)
print(dist.shape)
plt.scatter(dist[:,0],dist[:,1])
```
